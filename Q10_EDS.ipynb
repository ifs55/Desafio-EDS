```python
import matplotlib.pyplot as plt
from collections import Counter

# Exemplo de lista de datas
datas = ['2022-01-01', '2022-01-02', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-04', '2022-01-04']

# Conta a quantidade de atendimentos por dia
contagem = Counter(datas)

# Extrai as datas e as quantidades para criar o gráfico de barras
datas_grafico = list(contagem.keys())
quantidades_grafico = list(contagem.values())

# Cria o gráfico de barras
plt.bar(datas_grafico, quantidades_grafico)

# Adiciona rótulos e título ao gráfico
plt.xlabel('Data')
plt.ylabel('Quantidade de atendimentos')
plt.title('Quantidade de atendimentos por dia')

# Mostra o gráfico
plt.show()
```
